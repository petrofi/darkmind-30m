## Download Checkpoint & Tokenizer
We will download a pre-trained checkpoint from Hugging Face Hub if available, or you can run inference using the locally available structure.

```python
import torch
from model.gpt import GPTLanguageModel, GPTConfig
from tokenizer.darkmind_tokenizer.tokenizer import Tokenizer # Adjust based on actual tokenizer structure

# Note: In a real scenario, you would load the weights from checkpoints/darkmind_30m.pt here.
# For demonstration purposes, we are loading the architecture to show how easy it is.

print("Loading DarkMind-30M Configuration...")
config = GPTConfig(
    vocab_size=32000,
    block_size=128,
    n_layer=4,
    n_head=4,
    n_embd=256,
    dropout=0.1
)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = GPTLanguageModel(config).to(device)
print(f"Model initialized on {device}!")
print(f"Total parameters: {sum(p.numel() for p in model.parameters()) / 1e6:.2f} M")
```

## Generate Text
Let's pass a dummy context and generate a sequence.

```python
# Dummy generation (without actual loaded weights, this will output random tokens)
context = torch.zeros((1, 1), dtype=torch.long, device=device)
generated_tokens = model.generate(context, max_new_tokens=50)
print(generated_tokens.tolist())
```
